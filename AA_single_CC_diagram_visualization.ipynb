```python
import matplotlib.pyplot as plt
import canonical_computations_package as CC
import itertools
import numpy as np
```

```python
# Wolfram code rule nr
rule = 110
CA_size = 10
T = 100
figsize_multiplier = 2

# Input set = {0,1}^2
input_set = np.array([list(input_example) for  input_example in itertools.product([0,1],repeat= 2)])

encoded_input_set = CC.ReCA.spatial_encoders.k_to_n_encoder(
                                                            input_set = input_set, 
                                                            encoded_input_size= CA_size, 
                                                            locations = [0,5]
    )

space_time_diagrams = []

figsize = (CA_size*figsize_multiplier,T*figsize_multiplier)
for i in range(encoded_input_set.shape[0]):
    initial_configuration = encoded_input_set[i,:]
    space_time_diagrams.append(CC.CA.ECA.run_ECA(110, initial_configuration, 100))
    
    plt.figure(figsize=figsize)
    plt.title("Rule: " + str(rule) + ", input: " + str(input_set[i]))
    plt.ylabel("Space")
    plt.xlabel("Time")
    plt.imshow(space_time_diagrams[-1])
    plt.show()
    
space_time_diagrams = np.array(space_time_diagrams)
```

```python
CC_labels = CC.diagrams.create_CC_labels(len(input_set))
CC_diagram = CC.diagrams.CC_diagram_consistent_labels(space_time_diagrams, CC_labels)
plt.figure(figsize=figsize)
plt.title("CC_diagram, rule: " + str(rule))
plt.ylabel("Space")
plt.xlabel("Time")
plt.imshow(CC_diagram)
```
